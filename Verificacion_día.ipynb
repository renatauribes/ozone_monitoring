```python
import pandas as pd
from openpyxl import load_workbook
```

```python
def cleaner(data, name):
    data["Fecha"] = pd.to_datetime(data["Fecha"], dayfirst=True)
    filtered = data.groupby(by=[data["Fecha"].dt.year, data["Fecha"].dt.month, data["Fecha"].dt.day]).count()
    filtered.to_csv(f"{name}.csv")
```

```python
wb = load_workbook('meteo_dn.xlsx')

for sheet in wb:
    # String of sheet title
    where = sheet.title
    ws = wb[where]
    # Read each sheet separately
    data = pd.read_excel('meteo_dn.xlsx', sheet_name=where)
    cleaner(data, where)
```
